```julia
using ProgressBars
using JLD2
using SparseArrays
using PyCall
using PyPlot
using NCDatasets
using Statistics

cartopy = pyimport("cartopy")
ccrs = pyimport("cartopy.crs")
cfeat = pyimport("cartopy.feature");
```

```julia
#2x2.5 resolution grid

ds_grid_2x25 = NCDataset("regrid_files/regular_lat_lon_91x144.nc")
area_2x25 = float.(ds_grid_2x25["area"][:]); #m2

lon = ds_grid_2x25["lon"][:]
lat = ds_grid_2x25["lat"][:]
;
```

```julia
#global population file in 2x2.5
ds_pop = NCDataset("/home/safrin/Python/Regrid_Output/Population_2015_2x2.5_PC_DC_global.nc");
pop_array = float.(ds_pop["pop"][:,:,1]); # (144,91) array
```

```julia
include("ACAI.jl")
```

### NOx sensitivity arrays ("fixed")

```julia
n_case = 46
n_year = 28

for i in 1:n_case
    ds = NCDataset("data/sensitivities/NOx/sensitivity_"*string(i)*".nc")
    RF = ds["RF"][:]
    PM25 = ds["PM25"][:]
    Sfc_O3 = ds["Sfc_O3"][:]
    Col_O3 = ds["Col_O3"][:]
    Zonal_O3 = ds["Zonal_O3"][:]
    mask = ds["mask"][:]
    lev_47 = ds["lev_47"][:]
    lev_72 = ds["lev_72"][:]

    sensitivity_dict[i] = SensitivityRegionImpacts(RF, PM25, Sfc_O3, Col_O3, Zonal_O3, mask, lev_47, lev_72)
end

NOx_sens = EmissionSensitivities(n_case, sensitivity_dict);
;
```

# AEIC2035 emissions

```julia
case = "AEIC2035"
```

```julia
layers = 72
ds = NCDataset("/home/joonheek/emission_inventories/AEIC2035/AEIC2035_agg_91x144.nc")
```

```julia
fb = float.(ds["FUELBURN"][:])
println("Fuel burn total [Tg] = ", round(sum(fb .* area_2x25)/1e9,digits=4))
```

```julia
nox = float.(ds["NO"][:].*(46/30) .+ ds["NO2"][:] .+ ds["HONO"][:].*(46/47))
println("NOx total [Gg] = ", round(sum(nox .* area_2x25)/1e6,digits=4))
println("NOx above 8km [Gg] = ", round(sum(nox[:,:,27:end] .* area_2x25)/1e6,digits=4))
```

```julia
figure(figsize=(10, 8))
ax = subplot(projection=ccrs.PlateCarree())

ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS)
h = ax.pcolormesh(repeat(lon,1,91), repeat(reshape(lat,1,length(lat)),144,1), 
                sum(nox,dims=3)[:,:,1].*area_2x25/1e6, transform=ccrs.PlateCarree(), 
                vmin = -4, vmax= 4, cmap="RdBu_r",
)

# title("AEIC2019 NOx emissions")
cbar = plt.colorbar(h,fraction=0.024, pad=0.01)
cbar.set_label("Gg NOx")
```

```julia
RF_nox = zeros(n_year)
PM25_nox = zeros(144,91,n_year)
Sfc_O3_nox = zeros(144,91,n_year)
Col_O3_nox = zeros(144,91,n_year)

multipliers = zeros(n_case)

for i in 1:n_case
# for i in 7:n_case
    mask = sensitivity_dict[i].mask
    if layers == 47
        lev = sensitivity_dict[i].lev_47
    elseif layers == 72
        lev = sensitivity_dict[i].lev_72
    end
    
    nox_temp = try nox[:,:,lev]
    catch
        0
    end
    
    multiplier = sum(nox_temp .* area_2x25 .* mask) * 1e-6 #Gg NOx    
    
    RF_nox .+= multiplier .* sensitivity_dict[i].RF
    PM25_nox .+= multiplier .* sensitivity_dict[i].PM25
    Sfc_O3_nox .+= multiplier .* sensitivity_dict[i].Sfc_O3
    Col_O3_nox .+= multiplier .* sensitivity_dict[i].Col_O3
    
    multipliers[i] = multiplier
end
```

### RF

```julia
RF_nox[1] .* 1e3 #mW/m2
```

```julia
RF_nox[1] .* 1e3 / (sum(nox .* area_2x25)/1e9)
```

### PM2.5

```julia
figure(figsize=(10, 8))
ax = subplot(projection=ccrs.PlateCarree())

ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS)
h = ax.pcolormesh(repeat(lon,1,91), repeat(reshape(lat,1,length(lat)),144,1), 
                PM25_nox[:,:,1], transform=ccrs.PlateCarree(), 
                vmin = -0.5, vmax= 0.5, cmap="RdBu_r",
)

title("Change in PM2.5 exposure from "*case)
cbar = plt.colorbar(h,fraction=0.025, pad=0.01)
cbar.set_label("μg/m³", labelpad=10)
```

```julia
figure(figsize=(10, 8))
ax = subplot(projection=ccrs.PlateCarree())

ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS)
h = ax.pcolormesh(repeat(lon,1,91), repeat(reshape(lat,1,length(lat)),144,1), 
                PM25_nox[:,:,1]./ (sum(nox .* area_2x25)/1e9), transform=ccrs.PlateCarree(), 
                vmin = -0.1, vmax= 0.1, cmap="RdBu_r",
)

cbar = plt.colorbar(h,fraction=0.025, pad=0.01)
cbar.set_label("μg/m³ per Tg NOₓ", labelpad=10)
```

```julia
sum(PM25_nox[:,:,1] .* pop_array) / sum(pop_array)
```

```julia
sum(PM25_nox[:,:,1] .* pop_array) / sum(pop_array) / (sum(nox .* area_2x25)/1e9)
```

```julia
figure(figsize=(10, 8))
ax = subplot(projection=ccrs.PlateCarree())

ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS)
h = ax.pcolormesh(repeat(lon,1,91), repeat(reshape(lat,1,length(lat)),144,1), 
                Sfc_O3_nox[:,:,1], transform=ccrs.PlateCarree(), 
                vmin = -3, vmax= 3, cmap="RdBu_r",
)

title("Change in surface O3 exposure from "*case)
cbar = plt.colorbar(h,fraction=0.025, pad=0.01)

cbar.set_label("ppbv", labelpad=10)
```
